## 1. Periodicidad de señales en el tiempo

```python
import pandas as pd
import numpy as np
import os
from scipy import signal
```

```python
# Ruta de la carpeta de señal
carpeta_senal = 'Señal'

# Obtener todos los archivos CSV de la carpeta
archivos_csv = [f for f in os.listdir(carpeta_senal) if f.endswith('.csv')]

# Ordenar los archivos
archivos_csv.sort()

# Función para determinar si una señal es periódica usando autocorrelación
def es_periodica(senal, umbral=0.5):
    # Calcular autocorrelación
    autocorr = np.correlate(senal, senal, mode='full')
    autocorr = autocorr[len(autocorr)//2:]  # Tomar solo la mitad positiva
    autocorr = autocorr / autocorr[0]  # Normalizar
    
    # Buscar picos en la autocorrelación (excluyendo el pico en lag=0)
    picos, _ = signal.find_peaks(autocorr[1:], height=umbral)
    
    return len(picos) > 0, autocorr

# Procesar cada archivo CSV
for archivo_csv in archivos_csv:
    ruta_completa = os.path.join(carpeta_senal, archivo_csv)
    
    # Leer el CSV y tomar las filas 3 a 1003 (índices 2 a 1002)
    df = pd.read_csv(ruta_completa)
    datos_utiles = df.iloc[3:1003]
    
    # Asumiendo que la señal está en la primera columna
    # Convertir a numérico, forzando errores a NaN
    senal = pd.to_numeric(datos_utiles.iloc[:, 0], errors='coerce').values
    
    # Eliminar valores NaN si existen
    senal = senal[~np.isnan(senal)]
    
    # Determinar si es periódica
    periodica, autocorr = es_periodica(senal)
    
    print(f"{archivo_csv}: {'Periódica' if periodica else 'No periódica'}")
```

## 2. Ancho de espectro

```python
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Ruta de la carpeta de espectro
carpeta_espectro = Path('Espectro')

# Verifica que la carpeta existe
if not carpeta_espectro.exists():
    raise FileNotFoundError(f"Carpeta no encontrada: {carpeta_espectro}")

# Obtener lista de todos los archivos CSV en la carpeta Espectro
archivos_espectros = sorted(carpeta_espectro.glob('*.csv'))
print(f"Total de archivos encontrados: {len(archivos_espectros)}")

def cargar_espectro(archivo):
    """
    Carga un archivo CSV de espectro (solo Traza A)
    Retorna un DataFrame con Longitud de Onda y Nivel
    """
    # Leer el CSV desde la línea 245 donde están los datos
    df = pd.read_csv(archivo, skiprows=244)
    
    # Limpiar nombres de columnas (eliminar espacios)
    df.columns = df.columns.str.strip()
    
    # Extraer solo las columnas de la Traza A
    espectro = pd.DataFrame({
        'Longitud_Onda_nm': pd.to_numeric(df.iloc[:, 0], errors='coerce'),
        'Nivel': pd.to_numeric(df.iloc[:, 1], errors='coerce')
    })
    
    # Eliminar filas con NaN
    espectro = espectro.dropna()
    
    return espectro

def aplicar_media_movil(datos, ventana=5):
    """
    Aplica un filtro de media móvil de n puntos
    """
    return pd.Series(datos).rolling(window=ventana, center=True).mean().values

def normalizar_espectro(datos):
    """
    Normaliza los datos entre 0 y 1
    """
    min_val = datos.min()
    max_val = datos.max()
    return (datos - min_val) / (max_val - min_val)

def graficar_espectro(df, titulo='Espectro', lambda_picos=None):
    """
    Grafica un espectro
    """
    plt.figure(figsize=(12, 6))
    plt.plot(df['Longitud_Onda_nm'], df['Nivel'], linewidth=0.8)
    plt.xlabel('Longitud de Onda (nm)', fontsize=12)
    plt.ylabel('Nivel', fontsize=12)
    plt.title(titulo, fontsize=14)
    plt.grid(True, alpha=0.3)
    
        # Dibujar líneas verticales en cada pico
    if lambda_picos is not None:
        for lp in lambda_picos:
            plt.axvline(x=lp, color='r', linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.show()

def graficar_multiples_espectros(archivos, num_espectros=5, inicio=0):
    """
    Grafica múltiples espectros uno a uno en escala logarítmica y normalizada
    
    Args:
        archivos: lista de archivos a graficar
        num_espectros: número de espectros a graficar
        inicio: índice de inicio para seleccionar archivos
    """
    archivos_seleccionados = archivos[inicio:inicio + num_espectros]
    
    for archivo in archivos_seleccionados:
        df = cargar_espectro(archivo)
        df_filtrado = df

        # Convertir potencia a escala ogarítmica (dB)
        df_log = df_filtrado.copy()
        
        # Filtrar valores <= 0 antes de aplicar log10
        potencia = df_filtrado.iloc[:, 1]
        # Reemplazar valores <= 0 con un valor muy pequeño o eliminarlos
        potencia_filtrada = potencia.where(potencia > 0, np.nan)
        
        # Aplicar log10 solo a valores positivos
        df_log.iloc[:, 1] = 10 * np.log10(potencia_filtrada)
        
        # Eliminar filas con NaN generados por valores no positivos
        df_log = df_log.dropna()
        
        # Aplicar filtro de media móvil de 5 puntos
        df_log['Nivel'] = aplicar_media_movil(df_log['Nivel'].values, ventana=5)
        
        # Eliminar NaN generados por el filtro
        df_log = df_log.dropna()
        
        # Normalizar el espectro entre 0 y 1
        df_log['Nivel'] = normalizar_espectro(df_log['Nivel'].values)

        # Calcular la suma de todos los puntos del espectro normalizado
        suma_puntos = df_log['Nivel'].sum()

        # Definir umbral para determinar si se grafica
        umbral_suma = 125

        # Solo graficar si la suma supera el umbral
        if suma_puntos < umbral_suma:
            continue
        
        # Detectar picos en el espectro normalizado
        peaks = signal.find_peaks(df_log['Nivel'], height=0.4, prominence=0.15, distance=20)
        
        if len(peaks[0]) < 2:
            continue  # No hay picos, saltar a la siguiente iteración

        graficar_espectro(df_log, f'Espectro Normalizado: {archivo.name}', lambda_picos=df_log['Longitud_Onda_nm'].iloc[peaks[0]].values)
        print(f"Picos encontrados en {archivo.name} en longitudes de onda: {df_log['Longitud_Onda_nm'].iloc[peaks[0]].values}")

# Graficar todos los espectros normalizados
graficar_multiples_espectros(archivos_espectros, num_espectros=len(archivos_espectros), inicio=0)
```
